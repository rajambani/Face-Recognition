```python
#Positive class containing faces

import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
from array import array
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

directory = "C:\\Users\\Raj Ambani\\Desktop\\Fall 2018\\CS 584 - Machine Learning\\Project\\Dataset\\catdog\\"
i = -1

files = os.listdir(directory)
labels = np.zeros((10500))
features = []
#data = np.array((214, 256,256,1))

data = []
for filename in files:
    files = os.listdir(directory)  # Get all the files in that directory
    i = i+1
    print(i, filename)
    
    #Label creation
    name = filename.split('.')
    if(name[0].lower() == 'cat'):
        labels[i] = 0
    else:
        labels[i] = 1
    
    #Feature extraction
    image = Image.open(directory + filename)
    im = image.convert('L')
    im = im.resize((64,64))
    x = img_to_array(im)
    #print(x)
    x = x.ravel()
    print(i)
    X = np.array(x).reshape(x.shape)
    #print(X)
    data.append(X)

data = np.array(data)
data = data / 255
#print(positive_label)
print (data.shape, data)
print(labels.shape, labels)
```

```python
x_train_new, x_test_new, y_train_new, y_test_new = train_test_split(xnew, ynew, test_size=1./3,random_state = 6925,
                                                                        shuffle = True, stratify = ynew)
```

```python
# new model
new_model = Sequential()

new_model.add(Conv2D(32, kernel_size=(7,7),input_shape=(64,64,1),activation='relu'))
#new_model.add(Conv2D(32, kernel_size=(5,5),input_shape=(64,64,1),activation='relu'))
new_model.add(MaxPooling2D(pool_size=(2, 2)))
#new_model.add(MaxPooling2D(pool_size=(2, 2)))
new_model.add(Flatten())
new_model.add(Dropout(0.05))
#new_model.add(Dropout(0.05))
#new_model.add(Dense(1))
new_model.add(Dense(1,activation='sigmoid'))

new_model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])

epochs = 10
verbose = 1

for i in range(10,11):
    train_new_model = new_model.fit(x_train_new, y_train_new,epochs=i,verbose=verbose,
                                             validation_data=(x_test_new, y_test_new))

    #Save model
    new_model.save("init.model")

    y_predict_train_new = new_model.predict_classes(x_train_new)
    y_predict_test_new = new_model.predict_classes(x_test_new)

    print('i: ', i)
    f1_score_train_new = f1_score(y_train_new, y_predict_train_new)
    f1_score_test_new = f1_score(y_test_new, y_predict_test_new)

    print("f1_score_train_new: ", f1_score_train_new)
    print("f1_score_test_new: ", f1_score_test_new)
```
